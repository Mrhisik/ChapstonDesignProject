```python
import numpy as np
from keras.utils.np_utils import to_categorical
from keras.models import Sequential,Model
from keras.layers import Dense,Conv2D,Flatten,MaxPool2D,Input, BatchNormalization
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
import random
"""
class LeNet(Sequential):
    def __init__(self,num_classes):
        super().__init__()

        self.add(Conv2D(6,5,strides=1,activation='relu',input_shape=(32,32,1),padding='same'))
        self.add(AvgPool2D(2,strides=2))
        self.add(Conv2D(16,5,strides=1,activation='relu'))
        self.add(AvgPool2D(2,strides=2))
        self.add(Conv2D(120,5,strides=1,activation='relu'))
        self.add(Flatten())
        self.add(Dense(84,activation='tanh'))
        self.add(Dense(num_classes,activation='softmax'))

        self.compile(optimizer='adam',
                     loss='sparse_categorical_crossentropy',
                     metrics='accuracy')
model=LeNet(3)

model.summary()
"""
def Shortcut(layers):
    return layers[0] + layers[1]

def Conv(filters, kernel_size, strides, layer, activation=None):
    return tf.nn.leaky_relu(
        BatchNormalization()(
            Conv2D(filters, kernel_size, strides=strides, activation=tf.nn.leaky_relu if activation is None else activation, padding="same")(layer)
        )
    )

def ResidualBlock(filters, layer, activation=None):
    h = Conv(filters, (1, 1), 1, layer)
    h = Conv(filters, (3, 3), 1, h)
    h = Shortcut([layer, h])
    
    return Conv(filters, (1, 1), 1, h)

inputs = Input(shape=(32, 32, 1))
layer01 = Conv(32, 3, 1, inputs) # 16x16
resblk1 = ResidualBlock(32, layer01)
resblk1 = ResidualBlock(32, resblk1)
resblk1 = ResidualBlock(32, resblk1)

layer02 = Conv(64, 3, 2, resblk1) # 4x4
resblk2 = ResidualBlock(64, layer02)
resblk2 = ResidualBlock(64, resblk2)
resblk2 = ResidualBlock(64, resblk2)

layer04 = Flatten()(resblk2)
layer05 = Dense(1024, activation=tf.nn.leaky_relu)(layer04) # fully connection

out = Dense(6, activation=tf.nn.softmax)(layer05)

model = Model(inputs, out)
model.summary()

def my_seed_everywhere(seed: int = 42):
    random.seed(seed) # random
    np.random.seed(seed) # np
    os.environ["PYTHONHASHSEED"] = str(seed) # os
    tf.random.set_seed(seed) # tensorflow

my_seed = 42
my_seed_everywhere(my_seed)
```
